## Data Fetch from Zillow API

```python
from urllib.request import urlopen
import requests
import xml.etree.ElementTree as ET
import csv
```

#### Create the file and set the column names

```python
#set file name
download_dir = "HousingData_Zillow_set3.tsv"
#open file
csv = open(download_dir, "w")
#set coloumn name
columnTitleRow = "zpid \t street \t zipcode \t city \t state \t latitude \t longitude \t price \t bedrooms \t bathrooms \t finishedSqFt \t lotSizeSqFt \t yearBuilt \t yearUpdated \t numFloors \t view \t parkingType \t heatingSources \t heatingSystem \t rooms \t homeDescription \t neighborhood \t schoolDistrict \n" 
#zipcode, city, state, latitude, longitude, price, bedrooms, bathrooms, finishedSqFt,lotSizeSqFt,yearBuilt, yearUpdated, floors, view, parkingType, heatingSource, heatingSystem, appliances, floorCovering, rooms, description, neighborhood,schoolDistrict \n"
#write the column name
csv.write(columnTitleRow)
```

#### Set range for no. of pages to be scraped
#### Extract the data from API GetUpdatedPropertyDetails and store it to corresponding columns

```python
#set range for no. of pages to be scraped 
for i in range(40001003,40001006):
    a = str(i)
    #set url
    url = 'http://www.zillow.com/webservice/GetUpdatedPropertyDetails.htm?zws-id=YourZillowAPIKey&zpid=' + a 
    #open the xml file from the url
    tree = ET.parse(urlopen(url))
    #set root
    root = tree.getroot()
    
    elemList = []
    for elem in tree.iter():
          elemList.append(elem.tag)
    #print(elemList)
    
    #if message is successfully processed
    if root[1][1].text == '0':
        print(root[1][1].tag)
        
        #set row to empty string
        row = ""
        for variable in root.findall('response'):
            zpid = variable.find('zpid')
            if zpid != None:
                row = row + zpid.text + '\t'
            else:
                row = row + 'NA' + '\t'
                
            street = variable.find('address').find('street')   
            if street != None:
                row = row + street.text + '\t'
            else:
                row = row + 'NA' + '\t'

            zipcode = variable.find('address').find('zipcode')   
            if zipcode != None:
                row = row + zipcode.text + '\t'
            else:
                row = row + 'NA' + '\t'

            city = variable.find('address').find('city')   
            if city != None:
                row = row + city.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            state = variable.find('address').find('state')    
            if state != None:
                row = row + state.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            latitude = variable.find('address').find('latitude')    
            if latitude != None:
                row = row + latitude.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            longitude = variable.find('address').find('longitude')    
            if longitude != None:
                row = row + longitude.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            price = variable.find('price')
            if price != None:
                row = row + price.text + '\t'
            else:
                row = row + 'NA' + '\t'

            bedrooms = variable.find('editedFacts').find('bedrooms')    
            if bedrooms != None:
                row = row + bedrooms.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            bathrooms = variable.find('editedFacts').find('bathrooms')    
            if bathrooms != None:
                row = row + bathrooms.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            finishedSqFt = variable.find('editedFacts').find('finishedSqFt')    
            if finishedSqFt != None:
                row = row + finishedSqFt.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            lotSizeSqFt = variable.find('editedFacts').find('lotSizeSqFt')    
            if lotSizeSqFt != None:
                row = row + lotSizeSqFt.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            yearBuilt = variable.find('editedFacts').find('yearBuilt')    
            if yearBuilt != None:
                row = row + yearBuilt.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            yearUpdated = variable.find('editedFacts').find('yearUpdated')    
            if yearUpdated != None:
                row = row + yearUpdated.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            numFloors = variable.find('editedFacts').find('numFloors')    
            if numFloors != None:
                row = row + numFloors.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            view = variable.find('editedFacts').find('view')    
            if view != None:
                row = row + view.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            parkingType = variable.find('editedFacts').find('parkingType')    
            if parkingType != None:
                row = row + parkingType.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            heatingSources = variable.find('editedFacts').find('heatingSources')    
            if heatingSources != None:
                row = row + heatingSources.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            heatingSystem = variable.find('editedFacts').find('heatingSystem')    
            if heatingSystem != None:
                row = row + heatingSystem.text+ '\t'
            else:
                row = row + 'NA' + '\t'

            rooms = variable.find('editedFacts').find('rooms')    
            if rooms != None:
                row = row + rooms.text+ '\t'
            else:
                row = row + 'NA' + '\t'  

            neighborhood = variable.find('neighborhood')
            if neighborhood != None:
                row = row + neighborhood.text + '\t'
            else:
                row = row + 'NA' + '\t'

            schoolDistrict = variable.find('schoolDistrict')
            if schoolDistrict != None:
                row = row + schoolDistrict.text + '\t'
            else:
                row = row + 'NA' + '\t'
                
            homeDescription = variable.find('homeDescription')
            
            if homeDescription != None:
                row = row + homeDescription.text + '\n'
            else:
                row = row + 'NA' + '\n'

            csv.write(row)


    
#close csv
csv.close()    
```
